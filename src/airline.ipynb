### 1) Обоснование данных

```python
data = pd.read_csv('PassengerSatisfaction.csv', index_col=0)
data.head()
```

<p>Этот набор данных содержит опрос удовлетворенности авиапассажиров.<br>
Он позволяется нам проанализировать уровень комфорта авиакомпаний. <br>
Мы можем собрать статистику полетов и на её основании понять, в чем заключается комфорт человека в самолете.<br>
Эти знания помогут эффективней вкладывать деньги авиакомпаний и привлекать больше авиапассажиров.<br>
Можно задать много интересных вопросов по поводу эти данных, например: <br> -какой возраст,пол или класс полета чаще всего остается неудовлетворенным, <br>
-насколько сильно задержка рейса влияет на общее мнение о компании.<br>
Но самый главный вопрос:<br>
-Какие факторы связаны с удовлетворенным (или неудовлетворенным) пассажиром?</p>
Поэтому этот набор данных является очень полезным для авиакомпаний.

### 2) Элементарные статистики

<p>Посмотрим какой тип данных хранится в каждом признаке</p>

```python
data.dtypes
```

<p>Посмотрим нет ли пропущенных значений</p>

```python
data.isnull().sum()
```

<p>Удаляем пропущенные значения</p>

```python
data = data.dropna()
data.isnull().sum()
```

<p>Найдем категориальные и числовые признаки (часть индивидуального задания)</p>

```python
columns_numeric = data.select_dtypes(include=['float64', 'int64']).columns
columns_categorical = data.select_dtypes(include='object').columns
```

<p>Максимум</p>

```python
data[columns_numeric].max()
```

<p>Минимум</p>

```python
data[columns_numeric].min()
```

<p>Среднее значение</p>

```python
data[columns_numeric].mean()
```

<p>Медиана</p>

```python
data[columns_numeric].median()
```

<p>Standard deviation</p>

```python
data[columns_numeric].std()
```

<p>Также в pandas есть вспомогательная функция, с помощью которой можно одной строчкой подсчитать элементарные статистики</p>

```python
data.describe(include='all')
```

### 3) Распределение категориальных данных

```python
for col_categorical in columns_categorical:
    print(f'{data[col_categorical].value_counts()}\n')
```
